# **Linear Regression**

```
#Applying Linear Regression on Any Three Columns of Dataset except target Column Species

x=df.iloc[:,1:4].values
y=df.iloc[:,4].values


from sklearn.linear_model import LinearRegression #For performing linear Regression on the dataset

lin=LinearRegression()
lin.fit(x,y)
```

# **Prediction and Accuracy check of Linear Regression**

```
from sklearn.metrics import r2_score #For scoring

lin_preds = lin.predict(x)

print("R^2: ", r2_score(y, lin_preds))
```

# **Polynomial Regression**

```
# Fitting Polynomial Regression to the dataset
from sklearn.preprocessing import PolynomialFeatures #For performing Polynomial Regression on the dataset
x=df.iloc[:,1:4].values
y=df.iloc[:,4].values

poly = PolynomialFeatures(degree = 5)
x_poly = poly.fit_transform(x)

poly.fit(x_poly, y)
lin2 = LinearRegression()
lin2.fit(x_poly,y)
```

# **Prediction and Accuracy check of Polynomial Regression**

```
Poly_preds = lin2.predict(x_poly)
print("R^2: ", r2_score(y, Poly_preds))
```

# **Logistic Regression**

```
from sklearn.linear_model import LogisticRegression #For performing logistic Regression on the dataset
from sklearn.metrics import classification_report #Check precision, recall, f1-score
from sklearn.metrics import accuracy_score #This function compute sub-set accuracy
from sklearn.model_selection import train_test_split #For splitting dataset for training a model
```

```
# X = feature values, all the columns except the last column
logisx = df.iloc[:, 1:4].values

# y = target values, last column of the data frame
logisy = df.iloc[:,4].values
```

```
#Split the data into 70% training and 30% testing
x_train, x_test, y_train, y_test = train_test_split(logisx, logisy, test_size=0.3, random_state=0)
```

```
#Train the model
model = LogisticRegression()
model.fit(x_train, y_train) #Training the model
```

# **Prediction and Accuracy check of Logistic Regression**

```
#Test the model for Testing Data
predictions_test = model.predict(x_test)
print(predictions_test)# printing predictions

print()# Printing new line

#Check precision, recall, f1-score
print( classification_report(y_test, predictions_test) )

print( accuracy_score(y_test, predictions_test))
```

```
#Test the model for Training Data
predictions_train = model.predict(x_train)
print(predictions_train)# printing predictions

print()# Printing new line

#Check precision, recall, f1-score
print( classification_report(y_train, predictions_train) )

print( accuracy_score(y_train, predictions_train))
```

# **K Nearest Neighbors(KNN) and its Accuracy check**

```
from sklearn.preprocessing import StandardScaler # For scaling the data
from sklearn.model_selection import cross_val_score # For scoring
from sklearn.neighbors import KNeighborsClassifier # For applying KNN classification algorithm
from sklearn.ensemble import RandomForestClassifier # For applying Random Forest algorithm
```

```
#Applying Knn
knnws= KNeighborsClassifier(n_neighbors=10 , p=2 ,metric='minkowski')
knnws.fit(x_test,y_test)
knnws.score(x_test,y_test)
```

```
knns= KNeighborsClassifier(n_neighbors=10 , p=2 ,metric='minkowski')
knnws.fit(x_train,y_train)
knnws.score(x_train,y_train)
```

# **Scaling the Data**

```
sc = StandardScaler()
sc.fit(x_train)
X_train_std = sc.transform(x_train)
X_test_std = sc.transform(x_test)

#X_train_std and X_test_std are the scaled datasets to be used in algorithms
```

```
#Now Apply KNN on Scaled Data
knns.fit(X_test_std,y_test)
knns.score(X_test_std,y_test)
```

```
#Now Apply KNN on Scaled Data
knns.fit(X_train_std,y_train)
knns.score(X_train_std,y_train)
```

# **Random Forest**

```
#Applying Random Forest without Scaling Data
random_Forestws = RandomForestClassifier()
random_Forestws.fit(x_test,y_test)
random_Forestws.score(x_test,y_test)
```

```
random_Forestws.fit(x_train,y_train)
random_Forestws.score(x_train,y_train)
```

```
#Applying Random Forest With Scaling Data
random_Forests = RandomForestClassifier()
random_Forests.fit(X_test_std,y_test)
random_Forests.score(X_test_std,y_test)
```

```

random_Forests.fit(X_train_std,y_train)
random_Forests.score(X_train_std,y_train)
```
